```
file_location = "dbfs:/FileStore/tables/MOCK_DATA_BK_final.csv"
file_type = "csv"
infer_schema = "true" 
first_row_is_header = "true"   
delimiter = ";"
bd = spark.read.format(file_type) \
.option("inferSchema", infer_schema) \
.option("header", first_row_is_header) \
.option("sep", delimiter) \
.load(file_location)
bd.createOrReplaceTempView('MOCK_DATA_BK_final')
```

```
bd=bd.toPandas()
```

```
bd
```

```
y=bd.Sentimentos_Positivos
x=bd[['Preparo','Qtd_venda']]
```

```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
#treinamento
regressor.fit(x, y)
predictions =  regressor.predict(x)
predictions
bd=bd.copy()
bd['predictions']= predictions
#calculando o erro do modelo , absoluto e percentual
bd['error']=bd['Sentimentos_Positivos']-bd['predictions']
bd['erro_absoluto']=np.abs(bd['error'])
bd['erro_percen']=((bd['Sentimentos_Positivos']-bd['predictions'])/bd['Sentimentos_Positivos'])
bd['erro_absoluto']=np.abs(bd['error'])
bd['erro_percen']=((bd['Sentimentos_Positivos']-bd['predictions'])/bd['Sentimentos_Positivos'])
bd[['Preparo','Sentimentos_Positivos','predictions','error','erro_percen','erro_absoluto']].head(10)
#precisão do modelo
mae = 0
for i in range(0, len(predictions)):
    prediction = predictions[i]
    atual = y.iloc[i]
    error = abs(atual - prediction)
    mae = mae + error
    mae = mae / len(predictions)
print("mae",mae)
print("atual",atual.mean())
print("erro médio absoluto ( MAE ):",100*mae/atual.mean(),"%")
#coeficientes do modelo
intercept = regressor.intercept_
features = pd.DataFrame(regressor.coef_, x.columns, columns=['coefficient'])
features.head()
#print(features.importance_normalized)
#plotando grafico de feature importance
plt.bar(features.index, features.coefficient)
```
